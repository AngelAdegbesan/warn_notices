```python
###function to turn a dataframe into a csv
def turn_2_csv (name_of_file, data_frame):
    '''
    takes a dataframe and a file name and exports as a csv file
    '''
    print ("your file has been converted!!")
    return data_frame.to_csv(name_of_file, encoding= 'utf-8', index=False)
```

```python
##for datframe for more than $150k in loans
turn_2_csv ("more_than_150.csv", ny1_df)
```
