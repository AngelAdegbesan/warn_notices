```python
##targets the div that holds the table to scrape
organized = soup.find("div", class_="wysiwyg--field-webny-wysiwyg-body")
organized
```

```python
type(organized)
```

```python
##finds all a tags
ceos = organized.find_all ("a")
ceos
```

```python
type(ceos)
```

```python
##scrapes the url from the a tags and adds the base_url
co_urls = [item.get("href") for item in ceos]
base_url = "https://dol.ny.gov"
##for the base_urls that doesn't give an actual url, use request to get the fulll url and append to a larger list.
full_urls = []

for url in co_urls:
    if "https" not in url:            ##targets odd urls
        full_url = base_url + url 
    else:
        full_url = url
    if "pdf" not in full_url:
        res = requests.get(full_url)
        full_urls.append(res.url)
    else:
        full_urls.append(full_url)
full_urls
```

```python
len(full_urls)
```

```python
import os
print(os.getcwd())
```

```python
##importing libraries
##downloading the pdfs scraped
import shutil
import os

link_number = len(full_urls)
link_count = 1
##puts the newly downloaded files in its own folder
base_path = os.path.join(os.getcwd(), "2021_pdfs")
if not os.path.isdir(base_path):
    os.mkdir(base_path)

for link in full_urls:           
    ic(link)
    print (f"Downloading {link_count} of {link_number}")
    name = link.split("/")[-1]
    full_name = os.path.join(base_path, name)
    link_count += 1 

    ##requests and download
    sess = requests.session()
    with sess.get(link, stream=True) as resp:
        with open(full_name, "wb") as f:
            shutil.copyfileobj(resp.raw, f)
```

```python
##using the library to ocr all the pdfs into a single txt document.
ocr.ocr_files(myfiles, "multi.txt")
```

```python
##patterns to find information in the text.  #company, num_affect, class_pat, file_reason--full 220;get date&good
notice_date_pat = re.compile ("Date of Notice:\s([0-9]*\/[0-9]*\/[0-9]*)?(\w+\s\d{1,2},\s\d{4})?", re.I|re.M)
amend_date_pat = re.compile ("(Amendment:)\s([0-9]*\/[0-9]*\/[0-9]*)", re.I|re.M)
file_reason_pat = re.compile ("Reason Stated for Filing:\s([\w\s]*)\s", re.I|re.M)
company_pat = re.compile ("Company:\s([\w\s]*)", re.I|re.M)
num_affect_pat = re.compile ("Number Affected:\s([0-9]*)", re.I|re.M)
total_empl_pat = re.compile ("Total Employees:\s([0-9]*)", re.I|re.M)
business_type_pat = re.compile ("(Business Type:)\s([\w\s]*)\s", re.I|re.M)
disloc_reason_pat = re.compile ("Reason for Dislocation:\s([\w\s]*\W[\w\s]*)", re.I|re.M)
class_pat = re.compile ("Classification:\s([\w\s]*)\s", re.I|re.M)
```

```python
##open and read the single and newly created txt file
with open("output/multi.txt", 'r') as tdoc:
        file = tdoc.read()
```

```python
##puts the found patterns in a list
notice1 = list(notice_date_pat.findall(file))
company1 = list(company_pat.findall(file))
num_affect1 = list(num_affect_pat.findall(file))
classification1 = list(class_pat.findall(file))
file_reason1 = list(file_reason_pat.findall(file))
```

```python
business1 = list(business_type_pat.findall(file))
len(business1)
```

```python
##testing out patterns and lists to find missing data
notice1            ##company \n ##classification \n ## file_reaon \n ##
```

```python
##fill in the missing data on the notice_date_list and unpacking a tuple
date = []
for a,b, in notice1:
    date.append(a)
date[23] = "9/23/2021"
date
```

```python
##deleting unnecessary info
classification_lc = []
for item in classification1:
    item = item.replace("\n", "") 
    if not item == "":
        classification_lc.append(item)
```

```python
##deleting unnecessary info
file_reason_lc = []
for item in file_reason1:
    item = item.replace("\n", "") 
    if not item == "":
        file_reason_lc.append(item)
```

```python
##deleting unnecessary info
company_lc = []
for item in company1:
    item = item.replace("\n", "") 
    if not item == "":
        company_lc.append(item)
```

```python
## zip the variables and store in a list called txt_list
txt_list = []
for item in zip (company_lc, date, num_affect1, file_reason_lc, classification_lc):
    txt_list.append(item)  
```

```python
##sorts the list, alphabetically
txt_list_sort = natsorted(txt_list)
```

```python
##convert the list into a dataframe
df = pd.DataFrame (txt_list_sort)
df.columns = ["Company", "Notice Date", "Number Affected", "Reason", "Classification"]
df
```

```python
##convert the dateframe into a csv
file_name = "2021_warn_info.csv"
df.to_csv(file_name, encoding = "UTF-8", index = False)
```
